```python
#your code
woman_titanic["minor_wo"] = woman_titanic["Age"]<18
woman_titanic.groupby("minor_wo")["Age"].count()
```

16. How many woman that were less than 18 actually died? Call the variable `dead_wo`

```python
#your code


subdf = woman_titanic.loc[:,["minor_wo","Survived"]]

subdf["dead_wo"]= (subdf["minor_wo"]==True) & (subdf["Survived"]==0)
result = subdf.groupby("dead_wo").count()

print("In the titanic", result.iloc[1,0],"women less than 18 years old died")
```

24. Using [`.pivot_table()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html) to count how many women or men survived by class, call it `counted`.

Don't panic and read the documentation!




```python
titanic.groupby(["Sex", "Pclass"])["Survived"].count().unstack()
```

```python
#your code
counted = pd.pivot_table(titanic,values='Survived',index=["Pclass"],
                    columns=['Sex'], aggfunc="count")
counted
```
