# 04 - Grouping

### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                      sep='|', index_col='user_id')
users
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby("occupation")["age"].mean()
```

```python
users.groupby("occupation")[["age"]].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
#Agrupo por occupation y gender y cuento por gender
users2 = users.groupby(["occupation","gender"])["gender"].count()
```

```python
#pivoting the data frame
users3 = users2.unstack()
#umming both F + M in a new column total
users3["total"] = users3.apply(lambda x: x.sum(), axis =1)
#convert nan to 0
users3 = users3.fillna(0)
#creating a column with %male
users3["Male %"] = round(users3["M"]/users3["total"],2)
#sort
users3.sort_values('Male %', ascending = False)


```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby("occupation")["age"].aggregate([np.min,np.max])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(["occupation","gender"]).mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
users2 = users.groupby(["occupation", "gender"])["gender"].count()
users3 = users2.unstack()
users3.fillna(0, inplace = True)
users3["total"] = users3.apply(lambda x:x.sum(),axis =1)
users3["%M"] = round(users3["M"]/users3["total"],2)
users3["%F"] = round(users3["F"]/users3["total"],2)
```
