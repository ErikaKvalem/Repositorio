## Ejercicio 1
Somos una empresa que tiene varias tiendas de electrodomesticos. Necesitamos un programa para manejar las tiendas, ver las ventas que han tenido, cuántos empleados hay, etc... Para ello vamos a modelizar la tienda en una clase, que tendrá las siguientes características:
* Nombre clase: "Tienda"
* Atributos comunes:
    * Tipo: "Electrodomésticos"
    * Abierta: True
* Atributos propios de cada tienda:
    * Nombre - String
    * Dirección - String
    * Número de empleados - int
    * Ventas ultimos 3 meses - Lista de 3 numeros
* Método para calcular las ventas de todos los meses, que devuelva un numero con todas las ventas.
* Método que calcula la media de ventas de los ultimos meses, por empleado
* Método que devuelve en un string el nombre de la tienda, junto con su dirección.
* Método que obtiene las ventas del último mes.
* Método para dar una proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda. Siendo X el parámetro de entrada. Si X es menor de 1000, las ventas de los ultimos 3 meses hubiesen sido de (1.2 x ventas), si es mayor o igual de 1000, las ventas hubiesen sido de (1.5 x venta). El método reescribe el atributo de ventas en los últimos 3 meses, y además devuelve las ventas con el aumento.

Se pide:
1. Implementar la clase "Tienda"
2. Crear tres tiendas con datos inventados
3. Comprueba en al menos una de ellas, todo lo implementado en la clase tienda (sus atributos, media de ventas, ventas/empleado...)
3. Calcular las ventas del último mes de todas las tiendas. Para ello usa el bucle `for`
4. Imprime por pantalla los nombres de las tiendas cuya dirección lleve el string "Avenida"
5. Seguro que ya has hecho este apartado... Documenta la clase :)

```python
class Tienda:
    """
    Clase con los componentes de una tienda de Electrodomesticos S.S
    atributos:
    tipo = "Electrodomestico" - > String 
    abierta = True -> Bool
    
    """
    tipo = "Electrodomesticos"
    abierta = True
    
    def __init__(self,nombre,direccion,num_empleados, ventas_3_meses):
        """
        Elementos particulares de cada tienda
        Input:
            nombre : string
            direccion : string
            num_empleados : int
            ventas_3_meses : list
        """
        self.nombre = nombre #asocia el valor del objeto al propio objeto
        self.direccion = direccion
        self.num_empleados = num_empleados
        self.ventas_3_meses = ventas_3_meses
# Cada metodo va separado
    def total_ventas(self):
        """
        Metodo que suma todas las ventas de la tienda
        Output:
            total de ventas -> float
        """
        return sum(self.ventas_3_meses) #estamos sumando refiriendonos al atributo del objeto
     #sumando los elementos de la lista
    def ventas_empl(self):
        """
        Esto calcula la media de las ventas por empleado
            media de ventas -> float
            
        """
        return sum(self.ventas_3_meses)/self.num_empleados
        #return self.total_ventas / self.num_empleados
        #llamando al metodo que calcula al total del ventas
    
    def nombre_direc(self):
        """
        Metodo que devuelve un string con el nombre y la direccion
        Output
            nombre -> string
        """
        return "Nombre tienda:" + self.nombre + "\nDireccion:" + self.direccion
    def ventas_ult_mes(self):
        """
        Ventas de ultimo mes
        Output
            v_ult_mes -> float
        
        """
        v_ult_mes = self.ventas_3_meses[-1]
        return v_ult_mes
    def proyec_marketing(self,marketing):
        
        """
        Metodo para dar una proyeccion de las ventas
        Input: 
            marketing: cantidad invertida en marketing -> float
        Output: 
            lista de floats con el resultado del algoritmo
        Estos son los valores que deberian ser si la campaña de marketing a funcionado
        """
        if marketing < 1000:
            result1 = [1.2*n for n in  self.ventas_3_meses]
            self.ventas_3_meses = result1
            output = result1
        elif marketing >= 1000:
            result2 = [1.5*n for n in  self.ventas_3_meses]
            self.ventas_3_meses = result2
            output = result2
        return output
    
        
```

```python
# Apartado 2
tienda1 = Tienda("Mi tienda 1","Avenida del Electrodomestico 21", 3,[1000, 2000, 1500])
tienda2 = Tienda("Mi tienda 2","Avenida del Frigorifico 44", 5,[2000, 2000, 1500])
tienda3 = Tienda("Mi tienda 3","Camino de la tablet 10", 10,[10000, 6000, 12000])

# Apartado 3
print(tienda1.tipo)
print(tienda1.abierta)
print(tienda1.nombre)
print("\n")
print(tienda1.direccion)
print(tienda1.num_empleados)
print(tienda1.ventas_3_meses)
print(tienda1.total_ventas())
print(tienda1.ventas_empl())
print(tienda1.ventas_ult_mes())
print(tienda1.proyec_marketing(2000))


#Apartado 4
tot_tiendas = [tienda1, tienda2, tienda3]

total_ventas = 0

for tienda in tot_tiendas:
    total_ventas += tienda.ventas_ult_mes()
    
print(total_ventas)

# Apartado 5

for tienda in tot_tiendas:
    if "avenida" in tienda.direccion.lower():
        print(tienda.nombre)
    
```

## Ejercicio 2
En este ejercicio vamos a implementar una clase *Perro* en Python. La clase tiene las siguientes características:
* Cosas que sabemos seguro que tiene un perro
    * Tiene 4 patas
    * 2 orejas
    * 2 ojos
    * Una velocidad de 0. Por defecto, el perro está parado
* Cuando se inicialice:
    * El perro será de una determinada raza
    * Por defecto tendrá pelo "Marrón", a no ser que se diga lo contrario.
    * Por defecto no tendrá dueño, a no ser que se diga lo contrario.
    
* Dispondrá también de un método llamado andar, que tiene un argumento de entrada (aumento_velocidad). Este valor se le sumará a la velocidad que ya llevaba el perro.
* Necesita otro método (parar), donde pondremos la velocidad a 0.
* Otro método llamado "ladrar", que tendrá un argumento de entrada, y la salida será el siguiente string: "GUAU!" + el argumento de entrada.


Se pide:
* Implementa la clase *Perro*
* Crea un objeto de tipo *Perro*, sin dueño
* Comprueba que están bien todos sus atributos
* Prueba que ande, y comprueba su velocidad
* Páralo
* Documenta la clase *Perro*

```python
class Perro:
    """
    Clase perro de ejemplo
    Atributos:
        num_patas = 4 -> int 
        num_orejas = 2 -> int
        num_ojos = 2 -> int
        velocidad = 0.0 -> float
    """
    num_patas = 4
    num_orejas = 2
    num_ojos = 2 
    velocidad = 0.0
    
    def __init__(self,raza,pelo = "Marron", duenio = None):
        """
        Constructor para diferenciar un perro de otro
        Inputs: 
            raza -> string
            pelo -> string
            duenio -> string
        """
        
        self.raza = raza
        self.pelo = pelo
        self.duenio = duenio
    def andar(self,aumento_velocidad):
        """
        Aumentar el atributo velocidad
        Input:
            aumento_velocidad -> float
        """
        self.velocidad += aumento_velocidad
        
    def parar(self):
        """
        Poner el atributo velocidad a 0
        """
        self.velocidad = 0.0
    def ladrar(self,mensaje):
        """
        Metodo para mandar un mensaje
        Input: 
            mensaje -> string
        Output:
            string:
        """
        return"Guau!" + " " + mensaje
```

```python
bernardo = Perro("San Bernardo")
print(bernardo.raza)
print(bernardo.pelo)
print(bernardo.duenio)

bernardo.andar(2)
print(bernardo.velocidad)

bernardo.andar(2)
print(bernardo.velocidad)

bernardo.parar()
print(bernardo.velocidad)

bernardo.ladrar("Quiero comer")
```
